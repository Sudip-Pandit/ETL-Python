```
import pandas as pd
import numpy as np
```

```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder\
        .appName("ETL-PySpark")\
        .master("local[*]")\
        .getOrCreate()
```

```
spark
```

```
header = ['city', 'type', 'price']
data = map(lambda r: (r[0], r[1], float(r[2])),
map(lambda x: x.split(","),
    ["Nebraska, grocery, 23.66", "Ohio, food_items, 65.00", "NY, clothing_items, 6700,00",
    "California, diet_items, 765.00", "NJ, fast_food_items, 1200.00"]))
```

```
df = spark.createDataFrame(data, header)
```

```
df.show()
```

```
df.take(3)
```

```
df.printSchema()
```

```
#Selecting Columns
df.select("city").show()
```

```
df.select("type").show()
```

```
#How to use filter commands?
df.filter(df.city == 'Nebraska').show()
```

```
# How to do filter in multiple columns?
df.filter(df.city == "Nebraska").select(df.type == 'clothing_items').show()
```

```
df.filter(df.city == "Nebraska").select(df.type == 'grocery').show()
```

```
# How we can use filter function in two columns in the DataFrame?
df.filter(df.city == "Ohio").filter(df.type == 'food_items').show()
```

```
# We can use filter functions in multiple columsn as per our needs to use as a filter conditions?

df.filter(df.city == "Nebraska").filter(df.type == "grocery").filter(df.price < 25.0).show()
```

```
# We can use another way to filter the columns in pyspark

df.filter(
(df.city == "Nebraska") & (df.type > 20.0)).select("city").show()
```
